### Problem 2.9

```python
df['21_day_SMA'] = df['Close Price'].rolling(21).mean()
df['34_day_SMA'] = df['Close Price'].rolling(34).mean()
df
```

```python
df[['21_day_SMA', '34_day_SMA']].plot(figsize= (16,8))

plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

```python
df['Trade_Call'] = 0
for index, column in df.iterrows():
    if column['21_day_SMA'] != np.nan and column['34_day_SMA'] != np.nan:    
        if column['21_day_SMA'] < column['34_day_SMA']:
            df.at[index, 'Trade_Call'] = 0
        elif column['21_day_SMA'] > column['34_day_SMA']:
            df.at[index, 'Trade_Call'] = 1
        else:
            df.at[index, 'Trade_Call'] = 0
            
df['Trade_Call_Change'] = df['Trade_Call'].diff()
```

```python
df['Trade_Call'].value_counts()
```

```python
ax = df[['Close Price','21_day_SMA', '34_day_SMA']].plot(figsize= (16,8), color = ['black', 'blue', 'magenta'])


ax.plot(df['21_day_SMA'][df['Trade_Call_Change'] == -1].index,
        df['21_day_SMA'][df['Trade_Call_Change'] == -1], 'v', markersize=15, color = 'red', label = 'SELL')
ax.plot(df['21_day_SMA'][df['Trade_Call_Change'] == 1].index,
        df['21_day_SMA'][df['Trade_Call_Change'] == 1], '^', markersize=15, color = 'green', label= 'BUY')


plt.title('Trade Calls based on SMA of 21days and 34days')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

### Problem 2.10

```python
df["14_day_SMA"] = df['Close Price'].rolling(14).mean()
df["14_day_STD"] = df['Close Price'].rolling(14).std()
df["14_day_bollinger_upper"] = df["14_day_SMA"] + df["14_day_STD"] * 2
df["14_day_bollinger_lower"] = df["14_day_SMA"] - df["14_day_STD"] * 2
df.reset_index(inplace=True)

df.tail()
```

```python
ax = plt.gca()

df.plot(kind='line',figsize = (20,10), x='Date', y='Average Price', ax=ax, color='yellow')
df.plot(kind='line',figsize = (20,10), x='Date', y='14_day_SMA', ax=ax, color='cyan')
df.plot(kind='line',figsize = (20,10), x='Date', y='14_day_bollinger_upper', ax=ax, color='blue', linestyle='--')
df.plot(kind='line',figsize = (20,10), x='Date', y='14_day_bollinger_lower', ax=ax, color='red', linestyle='--')


plt.title("14 day bollinger bands")
plt.ylabel("Average Price")
plt.show()
```
