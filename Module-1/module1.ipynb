### Problem 1.5

```python
def average_priceNdays(N):
    return df['Close Price'].tail(N).mean()

def profitlossNdays(N):
    return (((df['Close Price'].tail(N).iloc[N-1] - df['Close Price'].tail(N).iloc[0])/df['Close Price'].tail(N).iloc[N-1]) * 100)
```

```python
print("The Average price over the last 1 week : ",average_priceNdays(7))
print("The profit/loss over the last week : ",profitlossNdays(7))
```

```python
print("The Average price over the last 2 week : ",average_priceNdays(14))
print("The profit/loss over the last 2 week : ",profitlossNdays(14))
```

```python
print("The Average price over the last 1 month : ",average_priceNdays(30))
print("The profit/loss over the last 1 month : ",profitlossNdays(30))
```

```python
print("The Average price over the last 3 months : ",average_priceNdays(90))
print("The profit/loss over the last 3 months : ",profitlossNdays(90))
```

```python
print("The Average price over the last 6 months : ",average_priceNdays(180))
print("The profit/loss over the last 6 months : ",profitlossNdays(180))
```

```python
print("The Average price over the last 1 year : ",average_priceNdays(365))
print("The profit/loss over the last 1 year : ",profitlossNdays(365))
```

### Problem 1.6

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change().fillna(0)
df.head()
```

### Problem 1.7

```python
if ((df['Day_Perc_Change'] >= -0.5) & (df['Day_Perc_Change'] <= 0.5)).all():
    df['Trend'] = 'Slight or No change'
if ((df['Day_Perc_Change'] >= 0.5) & (df['Day_Perc_Change'] <= 1)).all():
    df['Trend'] = 'Slight positive'
if ((df['Day_Perc_Change'] <= -0.5) & (df['Day_Perc_Change'] >= -1)).all():
    df['Trend'] = 'Slight negative'
if ((df['Day_Perc_Change'] >= 1) & (df['Day_Perc_Change'] <= 3)).all():
    df['Trend'] = 'Positive' 
if ((df['Day_Perc_Change'] <= -1) & (df['Day_Perc_Change'] >= -3)).all():
    df['Trend'] = 'Negative'
if ((df['Day_Perc_Change'] >= 3) & (df['Day_Perc_Change'] <= 7)).all():
    df['Trend'] = 'Among top gainers'
if ((df['Day_Perc_Change'] <= -3) & (df['Day_Perc_Change'] >= -7)).all():
    df['Trend'] = 'Among top losers'
if (df['Day_Perc_Change'] > 7).all():
    df['Trend'] = 'Bull run' 
if (df['Day_Perc_Change'] < -7).all():
    df['Trend'] = 'Bear drop'
    
df.head()
```

### Problem 1.8

```python
df_Trade = df[['Total Traded Quantity','Trend']]
print("Mean : \n",df_Trade.groupby(['Trend'])['Total Traded Quantity'].mean())
print("Median : \n",df_Trade.groupby(['Trend'])['Total Traded Quantity'].median())
```
