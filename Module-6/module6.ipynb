```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import scipy
```

### Problem 6.1

```python
files = [file for file in os.listdir() if file.endswith(".csv")]
print(files)
```

```python
df = pd.DataFrame()

for file in files:
    data = pd.read_csv(file)
    col_name = file.split(".")[0]
    df[col_name] = data['Close Price']
    
df
```

```python
df.info()
```

### Problem 6.2

```python
close_price = df.copy()
close_price = close_price.pct_change().dropna()

close_price
```

```python
returns = close_price.mean() * 252


volatility = close_price.std() * np.sqrt(252)
portfolio = pd.DataFrame(returns, columns=['returns'])
portfolio['volatility'] = volatility

portfolio
```

### Problem 6.3

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = np.array([portfolio['returns'].values, portfolio['volatility'].values])
x = x.T
x = scaler.fit_transform(x)
x
```

```python
distortions = []
K = range(1,10)
for k in K:
    model = KMeans(n_clusters=k).fit(x)
    distortions.append(sum(np.min(cdist(x, model.cluster_centers_, 'euclidean'), axis=1)) / x.shape[0])
    
    
print(distortions)
```

```python
plt.figure(figsize=(16,8))

plt.plot(K,distortions, 'bx-')
plt.title('Elbow Curve Method')
plt.xlabel('K')
plt.ylabel('Distortions')
```

##### so from the elbow method we can conclude that the optimum no of clusters is <b>4</b>

### Problem 6.4

```python
kmeans_model = KMeans(n_clusters=4).fit(x)
```

```python
portfolio['cluster'] = kmeans_model.predict(x)
portfolio
```

```python
portfolio['cluster'].replace({0 : "cluster1", 1 : "cluster2", 2 : "cluster3", 3 : "cluster4"}, inplace= True)
portfolio
```

```python
plt.figure(figsize=(15,8))

sns.scatterplot(x = portfolio['volatility'].values, 
                y = portfolio['returns'].values , 
                hue = portfolio['cluster'], 
                palette=['r','g','b','m'])

plt.title('Clusters of Stocks')
plt.xlabel('Volatility')
plt.ylabel('Returns')

plt.show()
```
